```python
installs_by_category = (
    df.groupby('category')['installs']
      .sum()
      .sort_values(ascending=False)
)
print("Installs by Category:\n", installs_by_category.head(10))
```

```python
installs_by_genre = (
    df.groupby('genres')['installs']
      .sum()
      .sort_values(ascending=False)
)
print("\nInstalls by Genre:\n", installs_by_genre.head(10))
```

```python
df_most_installed_categories = (
    df[df['installs'] > 1000000]   
      .groupby('category')           
      .size()                        
      .sort_values(ascending=False)  
      .head(5)                       
)
```

```python
print(f"The most installed categories are:\n{df_most_installed_categories}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_plot = df_most_installed_categories.reset_index()
```

```python
df_plot.columns = ['category', 'count']
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df_plot, x='count', y='category', palette='viridis')
```
